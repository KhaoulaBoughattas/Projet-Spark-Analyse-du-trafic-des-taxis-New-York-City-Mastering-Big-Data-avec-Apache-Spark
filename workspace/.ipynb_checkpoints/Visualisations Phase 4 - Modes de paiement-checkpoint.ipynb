```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as mtick

sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (12,6)
plt.rcParams['axes.titlesize'] = 16
plt.rcParams['axes.labelsize'] = 14

# Chargement des données
payment_by_trip_type = pd.read_csv("output/payment_by_trip_type/part-00000-a10c3553-f2ba-42f2-99fe-335c186d845c-c000.csv")
payment_over_time = pd.read_csv("output/payment_over_time/part-00000-8267303e-3440-4c9a-bf6f-66a6c9408459-c000.csv")
amount_by_payment = pd.read_csv("output/amount_by_payment/part-00000-9777733e-b102-44cc-929d-0089b60efb16-c000.csv")

#  Nombre de trajets par type de paiement (courts vs longs)
plt.figure(figsize=(10,6))
sns.barplot(x='trip_type', y='count', hue='payment_type', data=payment_by_trip_type)
plt.title("Nombre de trajets par type de paiement (courts vs longs)")
plt.xlabel("Type de trajet")
plt.ylabel("Nombre de trajets")
plt.legend(title="Mode de paiement")
plt.show()

#  Évolution des paiements par mois
payment_over_time['year_month'] = pd.to_datetime(payment_over_time['year_month'])
payment_pivot = payment_over_time.pivot(index='year_month', columns='payment_type', values='count')

plt.figure(figsize=(14,6))
payment_pivot.plot(marker='o', linewidth=2)
plt.title("Évolution du nombre de paiements par mois")
plt.xlabel("Mois")
plt.ylabel("Nombre de trajets")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()


#  Moyenne et médiane du montant total par mode de paiement
plt.figure(figsize=(10,6))
sns.barplot(x='payment_type', y='avg_total_amount', data=amount_by_payment, color='skyblue', label='Moyenne')
sns.barplot(x='payment_type', y='median_total_amount', data=amount_by_payment, color='orange', alpha=0.6, label='Médiane')
plt.title("Moyenne et médiane du montant total par mode de paiement")
plt.xlabel("Mode de paiement")
plt.ylabel("Montant total ($)")
plt.legend()
plt.show()

#  Boxplot du montant total par mode de paiement
plt.figure(figsize=(10,6))
sns.boxplot(x='payment_type', y='total_amount', data=pd.read_csv("output/total_amount_by_payment/part-00001-c8bff51d-7f8c-4c5d-a843-7ace9a830a26-c000.csv"))
plt.ylim(0, 200)  # Pour éviter que les outliers extrêmes écrasent le graphique
plt.title("Distribution du montant total par mode de paiement")
plt.xlabel("Mode de paiement")
plt.ylabel("Montant total ($)")
plt.show()

#  Analyse supplémentaire : proportion de paiement par type de trajet
payment_sum = payment_by_trip_type.groupby('trip_type')['count'].sum().reset_index()
payment_by_trip_type = payment_by_trip_type.merge(payment_sum, on='trip_type', suffixes=('', '_total'))
payment_by_trip_type['percentage'] = (payment_by_trip_type['count'] / payment_by_trip_type['count_total']) * 100

plt.figure(figsize=(10,6))
sns.barplot(x='trip_type', y='percentage', hue='payment_type', data=payment_by_trip_type)
plt.title("Proportion des modes de paiement selon le type de trajet")
plt.ylabel("Pourcentage (%)")
plt.xlabel("Type de trajet")
plt.legend(title="Mode de paiement")
plt.show()

#  Heatmap : évolution des paiements par mois et par type de paiement
payment_heatmap = payment_over_time.pivot(index='payment_type', columns='year_month', values='count')
plt.figure(figsize=(14,5))
sns.heatmap(payment_heatmap, annot=True, fmt=".0f", cmap="YlGnBu")
plt.title("Heatmap du nombre de paiements par mois et type de paiement")
plt.xlabel("Mois")
plt.ylabel("Mode de paiement")
plt.show()
```
